```python
with open('input_day6.txt', 'r') as f:
    inputs = f.readlines()
    groups = []
    lst = []
    i = 0
    for item in range(0, len(inputs)):
        if inputs[item] != '\n':
            inputs[item] = inputs[item].strip('\n')
            lst.append(inputs[item])
        if inputs[item] == '\n' or item == len(inputs)-1:
            groups.append(lst)
            lst = []
```

```python
def count_ans(group):
    set_yes = set()
    for person in group:
        for yes_ans in person:
            set_yes.add(yes_ans)
    return len(set_yes)
```

```python
def count_ans_unique(group):
    set_yes = set()
    string_letter = ''
    count_unique = 0
    
    if len(group) == 1:
        for yes_ans in group[0]:
            set_yes.add(yes_ans)
        count_unique = len(set_yes)
        
    elif len(group) > 1:
        for person in group:
            string_letter += person
            for yes_ans in person:
                set_yes.add(yes_ans)
        for lter in set_yes:
            if string_letter.count(lter) == len(group):
                count_unique += 1
                
    return count_unique
```

```python
count_yes = 0
for group in groups:
    count_yes += count_ans(group)
```

```python
count_unique = 0
for group in groups:
    count_unique += count_ans_unique(group)
count_unique
```
