```python
accumulate = 0
line = 0
instruction_list = []
```

```python
def operation(x):
    global accumulate, line, instruction_list
    x = x.split(' ')
    operation = x[0]
    argument = int(x[1])
    if operation == 'acc':
        accumulate += argument
        line += 1
    elif operation == 'jmp':
        line += argument
    elif operation == 'nop':
        line += 1

    if line not in instruction_list:
        instruction_list.append(line)
    else:
        line = 'a'
        print(accumulate)
```

```python
while True:
    operation(inputs[line])
```
